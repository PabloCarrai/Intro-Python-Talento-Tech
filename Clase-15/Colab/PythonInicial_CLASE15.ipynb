# 👥 Crear la base y tabla con validaciones

Comenzamos creando la base de datos `instituto.db` y la tabla `alumnos`, incluyendo restricciones que garanticen datos consistentes:

- `nombre`: obligatorio (no puede estar vacío)
- `dni`: obligatorio y único
- `edad`: debe ser un número entre 10 y 100

```python
import sqlite3

# Conexión y creación de tabla
conexion = sqlite3.connect("instituto.db")
cursor = conexion.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS alumnos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT NOT NULL,
    dni INTEGER NOT NULL UNIQUE,
    edad INTEGER CHECK(edad >= 10 AND edad <= 100)
)
""")

conexion.commit()
conexion.close()

print("Base de datos y tabla 'alumnos' creadas correctamente.")
```

# 🔍 ¿Qué hace este código?

📌 `NOT NULL` impide que el campo quede vacío.

📌 `UNIQUE` asegura que no haya dos alumnos con el mismo DNI.

📌 `CHECK` limita la edad a un rango razonable (10 a 100).

🎯 Esto hace que la base **proteja sus propios datos**, incluso si el programa tiene errores.


# 👥 Insertar alumnos iniciales (con validación)

Vamos a cargar algunos alumnos iniciales.  
Usamos parámetros seguros (`?`) y `try-except` para evitar que el programa se corte si hay datos repetidos.




```python
import sqlite3

alumnos = [
    ("Lucía Gómez", 30456789, 20),
    ("Ramiro Díaz", 27890123, 18),
    ("Elena Torres", 31800300, 22),
]

conexion = sqlite3.connect("instituto.db")
cursor = conexion.cursor()

for nombre, dni, edad in alumnos:
    try:
        cursor.execute("INSERT INTO alumnos (nombre, dni, edad) VALUES (?, ?, ?)", (nombre, dni, edad))
    except sqlite3.IntegrityError:
        print(f"Ya existe un alumno con el DNI {dni}. No se cargó.")

conexion.commit()
conexion.close()

print("Carga inicial completada.")
```

# 🔍 ¿Qué hace este código?

📌 `UNIQUE` impide duplicar DNIs  

📌 `try-except` evita que el programa se corte ante errores  


*   Elemento de la lista
*   Elemento de la lista


📌 Solo se insertan datos válidos

🎯 Probá ejecutar dos veces el bloque para ver qué ocurre con los DNIs duplicados.


# 👥 Consultar y mostrar todos los alumnos (ordenados por nombre)

Usamos `SELECT * FROM alumnos ORDER BY nombre` para obtener los datos en orden alfabético.

Este tipo de consulta es útil para mostrar listados en forma clara y organizada.


```python
import sqlite3

conexion = sqlite3.connect("instituto.db")
cursor = conexion.cursor()

cursor.execute("SELECT * FROM alumnos ORDER BY nombre")
alumnos = cursor.fetchall()

conexion.close()

print("Lista de alumnos:")
for id_, nombre, dni, edad in alumnos:
    print(f"{id_}. {nombre} – DNI: {dni} – Edad: {edad}")
```

# 🔍 ¿Qué hace este código?

📌 `ORDER BY nombre` organiza alfabéticamente el resultado.

📌 Se muestran todos los campos, incluyendo el ID autogenerado.

🎯 Este bloque será útil para verificar el estado actual de la base luego de cada operación.



# 👥 Agregar un nuevo alumno (con validación)

En este bloque vas a ingresar un nuevo alumno con `input()`.

Se valida que:
- El DNI no esté repetido
- La edad esté entre 10 y 100 años


```python
import sqlite3

try:
    nombre = input("Nombre completo: ").strip()
    dni = int(input("DNI: "))
    edad = int(input("Edad: "))

    if edad < 10 or edad > 100:
        print("Edad fuera de rango permitido (10–100).")
    else:
        conexion = sqlite3.connect("instituto.db")
        cursor = conexion.cursor()

        try:
            cursor.execute("INSERT INTO alumnos (nombre, dni, edad) VALUES (?, ?, ?)", (nombre, dni, edad))
            conexion.commit()
            print("Alumno agregado correctamente.")
        except sqlite3.IntegrityError:
            print("Ya existe un alumno con ese DNI.")

        conexion.close()

except ValueError:
    print("Ingresá solo números válidos para DNI y edad.")
```

# 🔍 ¿Qué hace este código?

📌 Validación de datos antes de insertarlos  

📌 Uso de `try-except` para controlar duplicados

📌 Inserción segura con parámetros (`?`)  

🎯 Probá ingresar datos válidos, un DNI repetido o una edad fuera de rango.


# 👥 Buscar alumno por DNI

Podés buscar un alumno por su DNI. Si existe, se muestra su información completa.

Si no se encuentra, se muestra un mensaje claro.


```python
import sqlite3

try:
    dni = int(input("Ingresá el DNI del alumno a buscar: "))

    conexion = sqlite3.connect("instituto.db")
    cursor = conexion.cursor()

    cursor.execute("SELECT * FROM alumnos WHERE dni = ?", (dni,))
    alumno = cursor.fetchone()

    conexion.close()

    if alumno:
        id_, nombre, dni, edad = alumno
        print(f"Alumno encontrado:\n{nombre} – DNI: {dni} – Edad: {edad}")
    else:
        print("No se encontró ningún alumno con ese DNI.")

except ValueError:
    print("El DNI debe ser un número.")
```

# 🔍 ¿Qué hace este código?

📌 `fetchone()` devuelve un solo registro, o `None` si no encuentra nada.

📌 Usamos `?` para evitar errores y proteger la consulta.

🎯 Probá buscar un DNI que exista y otro que no esté en la base.


# 👥 Modificar la edad de un alumno (por DNI)

Este bloque permite actualizar la edad de un alumno identificado por su DNI.

Se valida que:
- El nuevo valor esté entre 10 y 100
- El DNI exista en la base de datos


```python
import sqlite3

try:
    dni = int(input("DNI del alumno a modificar: "))
    nueva_edad = int(input("Nueva edad: "))

    if nueva_edad < 10 or nueva_edad > 100:
        print("Edad fuera del rango permitido (10–100).")
    else:
        conexion = sqlite3.connect("instituto.db")
        cursor = conexion.cursor()

        cursor.execute("UPDATE alumnos SET edad = ? WHERE dni = ?", (nueva_edad, dni))
        if cursor.rowcount > 0:
            conexion.commit()
            print("Edad actualizada correctamente.")
        else:
            print("No se encontró ningún alumno con ese DNI.")

        conexion.close()

except ValueError:
    print("Ingresá solo números válidos.")
```

# 🔍 ¿Qué hace este código?

📌 Validaciones numéricas para evitar errores de ingreso

📌 `cursor.rowcount` para saber si se modificó algún registro

📌 Protección de los datos: no se actualiza si el DNI no existe o la edad es inválida

🎯 Probá con distintos DNIs y edades, y verificá luego con el bloque de búsqueda.


# 👥 Eliminar un alumno (con confirmación)

Este bloque permite eliminar un alumno de la base por su DNI.

Antes de hacerlo, se pide confirmación al usuario.


```python
import sqlite3

try:
    dni = int(input("DNI del alumno a eliminar: "))
    confirmar = input(f"¿Estás seguro de eliminar al alumno con DNI {dni}? (s/n): ").strip().lower()

    if confirmar == "s":
        conexion = sqlite3.connect("instituto.db")
        cursor = conexion.cursor()

        cursor.execute("DELETE FROM alumnos WHERE dni = ?", (dni,))
        if cursor.rowcount > 0:
            conexion.commit()
            print("Alumno eliminado correctamente.")
        else:
            print("No se encontró ningún alumno con ese DNI.")

        conexion.close()
    else:
        print("Operación cancelada.")

except ValueError:
    print("Ingresá un número válido para el DNI.")
```

# 🔍 ¿Qué hace este código?

📌 Confirmación explícita antes de borrar

📌 Validación de entrada con `try-except`

📌 Se verifica si el alumno existía antes de intentar eliminarlo

🎯 Luego de eliminar, podés usar el bloque de búsqueda o listado para verificar los cambios.
